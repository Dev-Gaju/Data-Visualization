`ALso Known as box and whisker plot`

Box plots visually show the `distribution of numerical data and skewness` through displaying the data quartiles (or percentiles) and averages.

Box plots show the five-number summary of a set of data: including the minimum score, first (lower) quartile, median, third (upper) quartile, and maximum score.

If median is not in the middle then it could be define as 50% data below the median and 50% data above the median.Outliers define as point byond the whiskers.

* **Minimum Score** : Lowest score excluding outliers.
* **Lower Quartile** : Twenty five percent of score bellow the lower quartile.
* **Median : midpont**. diveded data into two part also known as 2nd quartile.
* **Upper quartile**: 75% of the scores fall below the upper quartile value and 25% above the value
* **maximum score**: Highest score without outliers.
* **Whisker** :The upper and lower whiskers represent scores outside the middle 50%
* **IQR** : 50% score from 25 to 75
